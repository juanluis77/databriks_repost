```

# Importar las bibliotecas necesarias,las bibliotecas necesarias de PySpark para poder trabajar con el modelo.
import matplotlib.pyplot as plt
import numpy as np
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.ml import Pipeline

```

```
# Crear una sesin de Spark
spark = SparkSession.builder.appName("RegresionLogistica").getOrCreate()

# Especificar la ruta al archivo
path = "/Users/201602892@est.umss.edu/tabla_minable_estandarizada.csv"

# Cargar el conjunto de datos desde un archivo CSV
data = spark.read.csv(path, header=True, inferSchema=True)
```

### AUC-ROC

```
# Seleccin de las columnas que vamos a usar para entrenar el modelo. Esto incluir las variables independientes y la variable dependiente.
feature_cols = [
    "loan_amount", "loan_duration", "loan_payments", "status_A", "status_B", 
    "status_C", "status_D", "type_DISPONENT", "type_OWNER", 
    "frequency_POPLATEK_MESICNE", "frequency_POPLATEK_PO_OBRATU", "frequency_POPLATEK_TYDNE"
]

# Seleccionar solo las columnas que necesitamos para el modelo
data = data.select(feature_cols + ['buen_candidato'])

# Dividir los datos en entrenamiento (80%) y prueba (20%)
train_df, test_df = data.randomSplit([0.8, 0.2], seed=1234)

# Configurar el ensamblaje de las caractersticas
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")

# Crear el modelo de regresin logstica
lr = LogisticRegression(featuresCol="features", labelCol="buen_candidato", maxIter=10)

# Crear el pipeline
pipeline = Pipeline(stages=[assembler, lr])

# Entrenar el modelo
model = pipeline.fit(train_df)

# Hacer predicciones sobre el conjunto de prueba
predictions = model.transform(test_df)

# Evaluador de clasificacin binaria para AUC-ROC
evaluator = BinaryClassificationEvaluator(labelCol="buen_candidato", metricName="areaUnderROC")
roc_auc = evaluator.evaluate(predictions)

# Imprimir AUC-ROC
print(f"\nAUC-ROC: {roc_auc}")
```

```
import numpy as np
import matplotlib.pyplot as plt
from pyspark.ml.linalg import Vectors
from sklearn.metrics import roc_curve, auc

# Obtener las probabilidades predichas (slo la segunda columna que contiene la probabilidad de la clase positiva)
y_pred_prob = predictions.select('probability').rdd.map(lambda row: row['probability'][1]).collect()

# Obtener las etiquetas reales
y_test = predictions.select('buen_candidato').rdd.map(lambda row: row['buen_candidato']).collect()

# Calcular la Curva ROC y AUC utilizando sklearn
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
roc_auc = auc(fpr, tpr)

# Graficar la curva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Lnea diagonal de no discriminacin
plt.xlabel('FPR (Tasa de Falsos Positivos)')
plt.ylabel('TPR (Tasa de Verdaderos Positivos)')
plt.title('Curva ROC')
plt.legend(loc='lower right')
plt.grid()
plt.show()

# Imprimir nuevamente el AUC-ROC
print(f"AUC-ROC calculado: {roc_auc}")

```

### Otras Mtricas (Matriz de Confusin, Reporte de Clasificacin, Log Loss)

```
# Para obtener el log_loss, necesitamos las probabilidades de prediccin
y_pred_prob = predictions.select('probability').rdd.map(lambda row: row['probability'][1]).collect()  # Probabilidades para AUC-ROC y log_loss
y_test = predictions.select('buen_candidato').rdd.map(lambda row: row['buen_candidato']).collect()

# Calcular el log_loss manualmente
from sklearn.metrics import log_loss
log_loss_value = log_loss(y_test, y_pred_prob)

# Matriz de confusin: podemos obtener el valor de la prediccin directamente en el DataFrame y compararlo
from sklearn.metrics import confusion_matrix
y_pred = predictions.select('prediction').rdd.map(lambda row: row['prediction']).collect()
conf_matrix = confusion_matrix(y_test, y_pred)

# Reporte de clasificacin: Obtn precisin, recall, F1 usando classification_report de sklearn
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)

# Imprimir resultados
print(f"\nMatriz de Confusin: \n{conf_matrix}")
print(f"\nReporte de Clasificacin:\n{report}")
print(f"\nBinary Cross-Entropy Loss (log_loss): {log_loss_value}")
```
